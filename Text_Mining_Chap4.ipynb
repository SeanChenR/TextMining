```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

import jieba.posseg as pseg
import pandas as pd
import gdown
import re
```

```python
dataset = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
    ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']
]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary , columns = te.columns_)
print('item : {}' .format(te.columns_))
print('matrix of first 5 samples : \n{}' .format(df.head(5)))
```

```python
frequent_itemsets = apriori(df , min_support = 0.001 , use_colnames = True)
print(frequent_itemsets)
print("-"*50)

frequent_itemsets = apriori(df , min_support = 0.5 , use_colnames = True)
print(frequent_itemsets)
print("-"*50)

frequent_itemsets = apriori(df , min_support = 0.7 , use_colnames = True)
print(frequent_itemsets)
print("-"*50)
```

```python
frequent_itemsets = apriori(df , min_support = 0.5 , use_colnames = True)

association_rules_by_conf = association_rules(frequent_itemsets , metric="confidence" , min_threshold=0.5)

association_rules_by_conf
```

```python
frequent_itemsets = apriori(df , min_support = 0.5 , use_colnames = True)

association_rules_by_conf = association_rules(frequent_itemsets , metric="confidence" , min_threshold=0.8)

association_rules_by_conf
```

```python
frequent_itemsets = apriori(df , min_support = 0.5 , use_colnames = True)

association_rules_by_conf = association_rules(frequent_itemsets , metric="lift" , min_threshold=1.2)

association_rules_by_conf
```

```python
frequent_itemsets = apriori(df , min_support = 0.5 , use_colnames = True)

ar_by_conf = association_rules(frequent_itemsets , metric="confidence" , min_threshold=0)

ar_by_conf
```

```python
ar_by_conf['antecedent_len'] = ar_by_conf['antecedents'].apply(lambda x : len(x))

result = ar_by_conf[(ar_by_conf['antecedent_len'] >= 2) & 
                                 (ar_by_conf['confidence'] >= 0.75) &
                                 (ar_by_conf['lift'] > 1.2)]

result
```

```python
url = 'https://drive.google.com/uc?id=1znE8-XebM5VZsQ-3Y9nVlXTAb55XWK9f'
output = 'new.json'
gdown.download(url , output , quiet=False)
data = pd.read_json(open('new.json' , 'r' , encoding = 'utf-8'))
data.head(2)
```

```python
url = 'https://drive.google.com/uc?id=1kdKO6DpI5p1dKnSwTDJVyA4Pr2Psw0WE'
gdown.download(url , 'stopword.txt' , quiet=False)
with open('stopword.txt' , 'r' , encoding='utf-8') as f:
    STOPWORDS = [word.rstrip("\n") for word in f]
print(STOPWORDS)
```

```python
def segment_remove_stopword(sent , pos_type=[]):
    sent = [word for (word,pos) in pseg.cut(sent) if pos in pos_type]
    sent = [re.sub('[0-9]' , '' , word) for word in sent]
    sent = [word for word in sent if word.strip() != '']
    return [word for word in sent if word not in STOPWORDS]

data['segment'] = data['標題'].apply(segment_remove_stopword , pos_type=['n' , 'v'])
data.head(5)
```

```python
TransEncoder = TransactionEncoder()
Tras_array = TransEncoder.fit(data['segment']).transform(data['segment'])
dataset_df = pd.DataFrame(Tras_array , columns = TransEncoder.columns_)
len(dataset_df)
```

```python
#return the column name of the items
frequent_items_name_result = apriori(dataset_df , min_support=0.001 , use_colnames=True)

#generate rules
association_rules_result = association_rules(frequent_items_name_result , min_threshold=0)

#compute number of item
association_rules_result['antecedent_len'] = association_rules_result['antecedents'].apply(lambda x : len(x))

association_rules_result['consequents_len'] = association_rules_result['consequents'].apply(lambda x : len(x))

association_rules_result
```

```python
result = association_rules_result[(association_rules_result['antecedent_len'] >= 3) &
                                                     (association_rules_result['consequents_len'] >= 1) &
                                                     (association_rules_result['confidence'] >= 0.5) &
                                                     (association_rules_result['lift'] >= 200)]
result
```
