# 3.2.1 載入Stanza套件

```python
zh_text = ['人工智慧亦稱智域、機器智慧，指由人製造出來的機器所表現出來的智慧。',
           '通常人工智慧是指透過普通電腦程式來呈現人類智慧的技術。',
           '中華郵政未來智慧物流服務，將取之大眾智慧，中華郵政帶給民眾更好的便利生活。']

en_text = ['In 1951, a board of directors was created and premises were rented on Hankou Street in downtown Taipei to set up the so-called Soochow Preparatory School.',
           'The school became the first private university in Taiwan.']
```

```python
import stanza

stanza.download('en')
stanza.download('zh-hant')
```

```python
#Initialize Chinese annotator
stanza_zh_annotator = stanza.Pipeline(lang='zh-hant',
                                      processors='tokenize,lemma,pos,depparse',
                                      use_gpu=True)

#Initialize English annotator
stanza_en_annotator = stanza.Pipeline(lang='en',
                                      processors='tokenize,lemma,pos,depparse',
                                      use_gpu=True)
```

# 3.2.2 斷詞、詞性標註、句法剖析

```python
def show(doc):
    for i,sent in enumerate(doc.sentences):
        print("Sentence {}\nINDEX\tTEXT\tLEMMA\tPOS\tDepRel\tHeadId" .format(i+1))
        for j,word in enumerate(sent.words):
            print("{}\t{}\t{}\t{}\t{}\t{}" .format(j+1, word.text, word.lemma, word.pos, word.deprel, word.head))
        print("")
```

```python
print('S_sents_annotated_ws_pos_deprel_zh ->')
S_sents_annotated_ws_pos_deprel_zh = [stanza_zh_annotator(sent) for sent in zh_text]
for sent in S_sents_annotated_ws_pos_deprel_zh:
    show(sent)
```

```python
print('S_sents_annotated_ws_pos_deprel_en ->')
S_sents_annotated_ws_pos_deprel_en = [stanza_en_annotator(sent) for sent in en_text]
for sent in S_sents_annotated_ws_pos_deprel_en:
    show(sent)
```

# 3.2.3 命名實體辨識(NER)

```python
#Initialize English annotator for NER
stanza_en_annotator = stanza.Pipeline(lang='en',
                                      processors='tokenize,ner',
                                      use_gpu=True)
```

```python
def show_ner(doc):
    for i,sent in enumerate(doc.sentences):
        print("Sentence {}\nINDEX\tTEXT\tNER" .format(i+1))
        for j,token in enumerate(sent.tokens):
            print("{}\t{}\t{}" .format(j, token.text, token.ner))
        print("")
```

```python
print('S_sents_annotated_ner_en ->')
S_sents_annotated_ner_en = [stanza_en_annotator(sent) for sent in en_text]
for sent in S_sents_annotated_ner_en:
    show_ner(sent)
```
